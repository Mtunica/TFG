```python
import numpy as np
```

```python
time = []
theta1 = []
theta2 = []
theta3 = []

with open("analytical.txt","r") as file:
    for line in file:
        line = line.split(',')
        time.append(line[0])
        theta1.append(line[1])
        theta2.append(line[2])
        theta3.append(line[3])
```

```python
theta = []
steps = 1000
for n in range(steps):
    theta.append([0, theta1[n], theta2[n], theta3[n]])     

N = 3     
delta_theta = []
for t in range(len(time)):
    delta = 0
    for i in range(1, N+1):
        for j in range(i+1, N+1):
            #print("i,j:",i,j)
            delta += (float(theta[t][i])-float(theta[t][j]))**2
    delta = (1/3)*delta
    delta_theta.append(delta)         
```

```python
import csv
with open("h.txt", "w") as file:
    writer = csv.writer(file)
    for n in range(len(time)):
        writer.writerow([time[n],delta_theta[n],np.log(delta_theta[n])])
```

```python
import matplotlib.pyplot as plt
plt.yscale("log")
plt.scatter(time,delta_theta)   
plt.xticks([0,500,1000])
plt.show()
```
